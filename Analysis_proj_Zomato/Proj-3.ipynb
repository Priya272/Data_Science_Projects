```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
df=pd.read_csv("Zomato.csv")
df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
df.drop(['url','address','votes','phone','reviews_list', 'menu_item'],axis=1,inplace=True)
df.head(5)
```

```python
df.isna().sum()
```

```python
df.drop_duplicates(inplace=True)
df.head(5)
```

```python
df.shape
```

```python
df['rate'].unique()
```

```python
def changerate(value):
    if (value=='NEW' or value=='-'):
        return np.nan
    else:
        value=str(value).split('/')
        value=value[0]
        return float(value)
    
df['rate']=df['rate'].apply(changerate)
df.head(10)
```

```python
df['rate'].fillna(df['rate'].mean(),inplace=True)
df['rate'].isna().sum()
```

```python
df.dropna(inplace=True)
df.isna().sum()
```

```python
df.rename(columns={'approx_cost(for two people)':'cost for two','listed_in(type)':'type','listed_in(city)':'city'},inplace=True)
df.head(5)
```

```python
df.drop(['location'],axis=1,inplace=True)
df.head(5)
```

```python
df['city'].unique()
```

```python
df['cost for two'].unique()
df.head(5)
```

```python
def modify(value):
    value=str(value)
    if ',' in value:
        value=value.replace(',','')
        return float(value)
    else:
        return float(value)
    
df['cost for two']=df['cost for two'].apply(modify)
df['cost for two'].unique()
```

```python
df['rest_type'].unique()
```

```python
df['rest_type'].value_counts()
```

```python
rest=df['rest_type'].value_counts(ascending=False)
rest
```

```python
rest_less=rest[rest<500]
rest_less
```

```python
def change(value):
    if(value in rest_less):
        return 'others'
    else:
        return value
df['rest_type']=df['rest_type'].apply(change)
df['rest_type'].value_counts()
    
```

```python
df['city'].value_counts()
```

```python
df['cuisines'].value_counts(ascending=False)
```

```python
cui=df['cuisines'].value_counts(ascending=False)
cui_less=cui[cui<100]
cui_less
```

```python
def change2(value):
    if value in cui_less:
        return 'OTHERS'
    else:
        return value
    
df['cuisines']=df['cuisines'].apply(change2)
df['cuisines'].value_counts()
```
